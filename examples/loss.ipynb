```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import spline
%matplotlib inline
```

```python
regression_loss=[2.3589,1.604,1.3256,1.1928,1.0848,0.9829,0.9125,0.8598,0.7543,0.7123]
```

```python
classification_loss=[0.5638,0.2943,0.2186,0.1819,0.1554,0.1382,0.1246,0.1151,0.1003,0.0954]
```

```python
epochs=range(len(regression_loss))
```

```python
plt.figure()
x = np.array(epochs)
y = np.array(regression_loss)
x_smooth = np.linspace(x.min(), x.max(), 500)
y_smooth = spline(x, y, x_smooth)
plt.plot(x_smooth, y_smooth, 'r-', label = 'Training')
plt.xlabel('epoch')
plt.ylabel('Regression Loss')
```

```python
plt.figure()
x = np.array(epochs)
y = np.array(classification_loss)
x_smooth = np.linspace(x.min(), x.max(), 500)
y_smooth = spline(x, y, x_smooth)
plt.plot(x_smooth, y_smooth, 'r-', label = 'Training')
plt.xlabel('epoch')
plt.ylabel('Classification Loss')
```
